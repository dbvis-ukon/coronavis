```python
import os
from datetime import datetime
import pandas as pd
from zeep import Client
import loadenv
import db
from db import get_connection
```

```python
def process_county(id):
    cur.execute(f"""
        SELECT timestamp, cases7_per_100k 
        FROM cases_per_county_and_day 
        WHERE county_id = '{id}'
        AND timestamp >= '2021-04-01' ORDER BY timestamp""")
    data = cur.fetchall()
    
    retData = []
    for d in data:
        retData.append({
            'id': id,
            'ts': d[0],
            'val': int(d[1]),
            'over100': False,
            'over165': False
        })
            
    for i in range(4, len(retData)):
                
        # check for date idx = i if ebrake has started
        # must be over t for 3 days
        retData[i]['over100'] = True
        retData[i]['over165'] = True
        for j in range(i - 4, i - 1):
            if data[j][1] < 165:
                retData[i]['over165'] = False
                
            if data[j][1] < 100:
                retData[i]['over100'] = False
                break
                
        # check the last 5 weekdays
        wasEBrake165 = False
        wasEBrake100 = False
        allUnder165 = True
        allUnder100 = True
        for j in range(max(0, i - 8), i - 2):
            if retData[j]['over165'] is True:
                wasEBrake165 = True
                
            if retData[j]['over100'] is True:
                wasEBrake100 = True
            
            # only workdays count
            if retData[j]['ts'].isoweekday() in [6,7]:
                continue
                
            if wasEBrake165 and retData[j]['val'] >= 165:
                allUnder165 = False
                
            if wasEBrake100 and retData[j]['val'] >= 100:
                allUnder100 = False
        
        if retData[i]['over100'] is False:
            retData[i]['over100'] = not allUnder100
            
        if retData[i]['over165'] is False:
            retData[i]['over165'] = not allUnder165
    
    for d in retData:
        print(d)
```

```python
conn, cur = get_connection()

cur.execute("SELECT county_id FROM cases_per_county_and_day GROUP BY county_id")
county_ids = cur.fetchall()


process_county('08335')
# for c in county_ids:
#     id = c[0]
#     print(id)
    
#     cur.execute(f"""
#         SELECT timestamp, cases7_per_100k 
#         FROM cases_per_county_and_day 
#         WHERE county_id = '{id}'
#         AND timestamp >= '2021-04-20' ORDER BY timestamp""")
#     data = cur.fetchall()
#     print(data)
    
```
